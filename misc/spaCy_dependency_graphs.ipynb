```python
import pandas as pd

import spacy
from spacy import displacy

import spacy as sp
nlp = sp.load('en_core_web_trf')

sentence = "Which of the following are cat or dog names?: Sylvester, Fido, Spot, Fluffy" #"Rats are various medium-sized, long-tailed rodents."
doc = nlp(sentence)
displacy.render(doc, style="dep")
```

```python
dir(doc)
```

```python
my_doc = doc.to_json()
dir(my_doc)
```

```python
nodes = pd.DataFrame(my_doc['tokens'])
nodes['token'] = [ my_doc['text'][ row['start'] : row['end'] ] for row in nodes.to_dict(orient='records') ]
nodes
```
